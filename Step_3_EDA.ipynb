```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pickle
%matplotlib inline
```

## EDA to investigate scope adjustment

```python
df = pd.read_pickle("autonom.p")
```

```python
df_months = df.groupby(pd.TimeGrouper(freq='M'))
```

```python
plt.plot(df_months.size())
plt.xlabel('time')
plt.xticks(rotation = 'vertical')
plt.ylabel('priority')

plt.show()
```

### Observations
1. Priority dates of patents on autonomous vehicles have dropped in the middle of 2016. This is probably due to provisional patent filings. A provisional patent is a placeholder-type of document before a full-fledged utility patent can be filed. Provisional patents remain unavailable to the public until and unless a utility patent is filed which claims priority to the filing of the provisonal patent.
2. Aside from the drop explained in 1., between years 2013 to 2016 patents have been filed at a pretty consistent rate each year.
