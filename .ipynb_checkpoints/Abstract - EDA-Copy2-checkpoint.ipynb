## Normalize Text

```python
def cleaner(text):

    text = BeautifulSoup(text, "html5lib").get_text()
    text = re.sub('&#39;',' ',text).lower()
    text = re.sub('<br />',' ',text)
    text = re.sub('<.*>.*</.*>',' ', text)
    text = re.sub('[\d]',' ',text)
    text = re.sub('[0-9]',' ',text)
    text = re.sub('[^a-z ]',' ',text)
    text = re.sub('\\ufeff', ' ', text)
    text = re.sub('[^a-z ]',' ',text)
    text = ' '.join(text.split())
    
    return text
```

```python
patent_df_2['clean_text'] = patent_df_2['text'].apply(cleaner)
```

```python
stemmer = PorterStemmer()

def stem_tokens(tokens, stemmer): 
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

def tokenize(text):
    tokens = word_tokenize(text)
    tokens = [i for i in tokens if len(i) > 2] 
    tokens = [i for i in tokens if i not in stop_list] 
    stems = stem_tokens(tokens, stemmer)
    return stems
```

```python
list_abstracts = patent_df_2['clean_text'].values.tolist()
```

```python
tokenized_sents = [tokenize(i) for i in list_abstracts]
```

```python
patent_df_2['token_sents'] = tokenized_sents
```

```python
patent_df_2['len_tokens'] = patent_df_2['token_sents'].apply(lambda x: len(x))
```

## Identify Topics

```python
vectorizer = CountVectorizer(tokenizer=tokenize, stop_words="english")
X_cv = vectorizer.fit_transform(patent_df_2['clean_text'])
```

```python
lda   = LatentDirichletAllocation(n_topics=10, learning_method="batch", random_state=42)
model_cv = lda.fit(X_cv)
```

```python
def display_topics(model, feature_names, no_words = 10, plot = False, plot_dim=(5,2)):
    
    topics_tokens = []
    
    for topic_idx, topic in enumerate(model.components_):

        topic = zip(feature_names, topic)
        topic = sorted(topic, key=lambda pair: pair[1])
        
        topic_words = [(token, counts)
                       for token, counts in topic[:-no_words - 1:-1]]
        
        topics_tokens.append(topic_words)
        
        if not plot:
            print ("Topic %d:" % (topic_idx))
            print (topic_words)
        
    
    if plot:
        
        plot_matrix = np.arange(10).reshape(5,2)
    
        fig, ax = plt.subplots(figsize=(10, 10), nrows=5, ncols=2)
        
        topics = [
            {key: value for key, value in topic} 
                  for topic in topics_tokens
        ]
        
        row = 0
        
        for topic_id, topic in enumerate(topics):
            
            column = (0 if topic_id % 2 == 0 else 1)
                
            chart = pd.DataFrame([topic]).iloc[0].sort_values(axis=0)
            chart.plot(kind="barh", title="Topic %d" % topic_id, ax=ax[row, column])
                
            row += 1 if column == 1 else 0
        
        plt.tight_layout()
            

display_topics(model_cv, vectorizer.get_feature_names(), no_words=10, plot=True)
```
